```python
##1. **Which countries experienced the largest increase in access to electricity (%) from 2000 to 2020?**
```

```python
access_df = df[['Entity', 'Year', 'Access to electricity (% of population)']]
pivot_df = access_df.pivot(index='Entity', columns='Year', values='Access to electricity (% of population)')
pivot_df = pivot_df.dropna(subset=[2000, 2020])
pivot_df['increase_percent'] = pivot_df[2020] - pivot_df[2000]
largest_increase = pivot_df.sort_values(by='increase_percent', ascending=False).head(10)
largest_increase.reset_index()
```

```python
#3. **List countries where access to clean fuels for cooking increased steadily (no dips) from 2010 to 2020.**
```

```python
df.head()
```

```python
cooking_df = df[df['Year'].between(2010, 2020)][['Entity', 'Year', 'Access to clean fuels for cooking']].dropna()
def is_steady_increase(group):
    return group['Access to clean fuels for cooking'].is_monotonic_increasing

steady_increase_countries = (
    cooking_df.groupby('Entity')
    .filter(is_steady_increase)
    .drop_duplicates(subset=['Entity'])[['Entity']]
    .reset_index(drop=True)
)

steady_increase_countries
```

```python
#4. **Identify countries with a consistent increase in renewable electricity capacity per capita over the last 15 years.**
```

```python
renewable_df = df[df['Year'].between(2005,2020)][["Entity","Year","Renewable-electricity-generating-capacity-per-capita"]].dropna()
def is_renewable_increase(group):
    return group['Renewable-electricity-generating-capacity-per-capita'].is_monotonic_increasing

steady_increase = (renewable_df.groupby('Entity').filter(is_renewable_increase).drop_duplicates(subset=['Entity'])[['Entity']].reset_index(drop =True))
steady_increase                   
```

```python
#**Identify countries with a consistent upward trend in renewable electricity capacity using `.diff()` or `.pct_change()` over years.**
```

```python
df['Capacity_Pct_Change'] = df.groupby('Entity')['Renewable-electricity-generating-capacity-per-capita'].pct_change()
def is_consistently_growing(group):
    return (group['Capacity_Pct_Change'].dropna() > 0).all()

upward_countries = df.groupby('Entity').filter(is_consistently_growing)
upward_countries['Entity'].unique()
```

```python
# **Which countries had the steepest 5-year rolling average increase in electricity access?*
```

```python
df1 = df.sort_values(['Entity', 'Year'])
```

```python
df['Rolling_avg'] = df1.groupby('Entity')['Access to electricity (% of population)'].transform(lambda x:x.rolling(window = 5, min_periods = 5).mean())
```

```python
df['Rolling_change'] = df.groupby('Entity')['Rolling_avg'].diff()
```

```python
max_rolling_increase = df.groupby('Entity')['Rolling_change'].max().reset_index()
```

```python
max_rolling_increase.sort_values('Rolling_change', ascending = False)
```
